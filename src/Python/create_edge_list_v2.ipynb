```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from itertools import combinations
```

```python
df = pd.read_csv('/content/sdglist_doc1_pg10.csv')
```

```python
df
```

```python
#loop through columns to find combination sum of every 2
```

```python
my_list = df['SDG'].tolist()
```

```python
my_list
```

```python
from itertools import combinations

sample_list = df['SDG']
list_combinations = list()
for n in range(len(sample_list) + 1):
    list_combinations += list(combinations(my_list, 2))
label_list = list_combinations
```

```python
len(label_list) 
```

```python
#remove duplicates (or can select the first n)

res = []
for i in label_list:
    if i not in res:
        res.append(i)
```

```python
len(res)  #cool that works
```

```python
edge = pd.DataFrame(res)
```

```python
edge  #!!! great it works, now need to add the sum of the "weight" column, and i have my edge list!!!
```
