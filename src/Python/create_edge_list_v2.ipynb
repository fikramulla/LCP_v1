```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from itertools import combinations
```

```python
df = pd.read_csv('/content/sdglist_doc1_pg10.csv')
```

```python
df
```

```python
#loop through columns to find combination sum of every 2
```

```python
my_list = df['SDG'].tolist()
```

```python
my_list
```

```python
sample_list = df['SDG']
list_combinations = list()
for n in range(len(sample_list) + 1):
    list_combinations += list(combinations(my_list, 2))
label_list = list_combinations
```

```python
len(label_list) 
```

```python
#remove duplicates (or can select the first n)

res = []
for i in label_list:
    if i not in res:
        res.append(i)
```

```python
len(res)  #cool that works
```

```python
edge = pd.DataFrame(res)
```

```python
#edge  #!!! great it works, now need to add the sum of the "weight" column, and i have my edge list!!!
```

```python
edge.columns = ['from', 'to']
```

```python
#it looks like it retained the order, so try the same on weight column
#trying weight column

sample_list = df['weight']
list_combinations = list()
for n in range(len(sample_list) + 1):
    list_combinations += list(combinations(sample_list, 2))
count_list = list_combinations

len(count_list) 
```

```python
#can not remove duplicates the same way
#try selecting first 120 items in list
count_list = count_list[0:120]
count = pd.DataFrame(count_list)
```

```python
count["sum"] = count.iloc[:, 0] + count.iloc[:, 1]
```

```python
#count  #now this count keeps counting single pairs as 1, should be 0(?)
```

```python
count["inertia"] = (count.iloc[:, 0] * count.iloc[:, 1]) / (count.iloc[:, 0] + count.iloc[:, 1])
count["inertia"] = count["inertia"].fillna(0)
```

```python
#count
```

```python
edge["sum"] = count["sum"]
edge["inertia"] = count["inertia"]
```
