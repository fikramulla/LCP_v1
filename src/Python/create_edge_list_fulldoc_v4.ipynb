```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from itertools import combinations
```

```python
docno = "10"   #increment this for doc number X or other filename

file_in = "/content/sdglist_doc" + docno + ".csv"
file_out = "/content/my_edge_withdocno_doc" + docno + ".csv"

df = pd.read_csv(file_in)
```

```python
my_list = df['SDG'].tolist()

sample_list = df['SDG']
list_combinations = list()
for n in range(len(sample_list) + 1):
    list_combinations += list(combinations(my_list, 2))
label_list = list_combinations
```

```python
#remove duplicates (or can select the first n)
res = []
for i in label_list:
    if i not in res:
        res.append(i)

len(res)    #should be 120
```

```python
edge = pd.DataFrame(res)    #great it works, now need to add the sum of the "weight" column, and i have my edge list!!!
edge.columns = ['from', 'to']
```

```python
#it looks like it retained the order, so try the same on weight column
#trying weight column

sample_list = df['weight']
list_combinations = list()
for n in range(len(sample_list) + 1):
    list_combinations += list(combinations(sample_list, 2))
count_list = list_combinations

#can not remove duplicates the same way
#selecting first 120 items in list since it retained the order
count_list = count_list[0:120]
count = pd.DataFrame(count_list)
```

```python
count["sum"] = count.iloc[:, 0] + count.iloc[:, 1]
count["inertia"] = (count.iloc[:, 0] * count.iloc[:, 1]) / (count.iloc[:, 0] + count.iloc[:, 1])
count["sum"] = count["sum"].fillna(0)
count["inertia"] = count["inertia"].fillna(0)
```

```python
edge["sum"] = count["sum"]
edge["inertia"] = round(count["inertia"], 2)
edge["docid"] = docno
```
